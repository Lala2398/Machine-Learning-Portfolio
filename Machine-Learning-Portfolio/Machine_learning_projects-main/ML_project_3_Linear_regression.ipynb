Linear Regression part

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
from sklearn.preprocessing import LabelEncoder
```

```python
y=data["Price"]
```

```python
x=data.drop("Price",axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    train_size=0.80,
    random_state=1)
```

```python
lr=LinearRegression()
```

```python
nr=LabelEncoder()
```

```python
data.dtypes
```

```python
data["CouncilArea"] = data['CouncilArea'].astype(str)
```

```python
data["CouncilArea"].value_counts
```

```python
data['Suburb'] = nr.fit_transform(data['Suburb'])
data["Address"] = nr.fit_transform(data["Address"])
data["Type"] = nr.fit_transform(data["Type"])
data["Method"] = nr.fit_transform(data["Method"])
data["SellerG"] = nr.fit_transform(data["SellerG"])
data["Date"] = nr.fit_transform(data["Date"])
data["CouncilArea"] = nr.fit_transform(data["CouncilArea"])
data["Regionname"] = nr.fit_transform(data["Regionname"])
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test).round(3)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rms = mean_squared_error(y_test, y_pred, squared=False)
```

```python
rms
```

```python
rmse = mean_squared_error(y_test, y_pred, squared=True)
```
